## Working with PDB

```python
from Bio.PDB import *
```

```python
# load PBD structure
pdbl = PDBList()
pdbl.retrieve_pdb_file('1GCN', pdir = '.', file_format = 'pdb')
```

```python
pdb_file = 'pdb1gcn.ent'

# create a parser
p = PDBParser(QUIET=True)

# parse various structure parts
structure = p.get_structure('protein', pdb_file)
model = structure[0]
chain = model['A']
```

The Structure object follows the so-called SMCRA (Structure/Model/Chain/Residue/Atom) architecture :

* A structure consists of models (crystal structures generally have only one model (with id 0), while NMR files usually have several models)
* A model consists of chains
* A chain consists of residues
* A residue consists of atoms

![picture](https://drive.google.com/uc?export=view&id=1KbDCusv26kT0Xb9IX2bzI5PC2eBUNYZm)



```python
print(structure)
print(model)
print(chain)
```

```python
# get atoms
atom_struct = structure.get_atoms()

atom_struct # <- this means we need to unpack this object
```

```python
# make a list an show first 10 elements
atom_lst = list(atom_struct)
print(len(atom_lst))
print(atom_lst[:10])
```

```python
# show coordinates for first 10 atoms
for atom in atom_lst[:10]:
    print(atom.get_coord())
```

```python
# get residues
residues = structure.get_residues()
residues
```

```python
# make a list an show first 10 elements
res_lst = list(residues)
print(len(res_lst))
print(res_lst[:10])
```

```python
# extracting specific parts
residue = chain[20]
atom = residue["CA"]

print(residue)
print(atom)
```

```python
# get residue properties

print(residue.get_resname()) # return the residue name (eg. 'GLY')
print(residue.is_disordered()) # 1 if the residue has disordered atoms
print(residue.get_segid()) # return the SEGID
print(residue.has_id("CA")) # test if a residue has a certain atom
print(residue.has_id("S"))
```

```python
# get atom properties

print(atom.get_name())  # atom name (spaces stripped, e.g. 'CA')
print(atom.get_id())  # id (equals atom name)
print(atom.get_coord())  # atomic coordinates
print(atom.get_vector())  # atomic coordinates as Vector object
print(atom.get_fullname())  # atom name (with spaces, e.g. '.CA.')
```

```python
# simple calculations

at1 = residue["CA"]
at2 = residue["CD"]
at3 = residue["N"]

vector1 = at1.get_vector()
vector2 = at2.get_vector()
vector3 = at3.get_vector()

distance = at1 - at2
angle = calc_angle(vector1, vector2, vector3)

print(distance)
print(angle)
```

### Working with FASTA

```python
from Bio import SeqIO
```

```python
filename='test_file.fasta'
ext = 'fasta'

sequences = SeqIO.parse(filename, ext) # seq input; returns an iterator
sequences
```

```python
for seq in sequences:
  print(seq)
```

```python
sequences = SeqIO.parse(filename, ext)
seq1 = next(sequences)
```

```python
print(seq1.id) # symbols before first whitespace
print(seq1.name) # same
print(seq1.description) # definition line
print(seq1.seq) # sequence line
```
